```python
df2=df1.drop(columns=['latitude','longitude'])
```

```python
# df2['listing_name'].fillna(, inplace=True)
# df2['host_name'].fillna(, inplace=True)
mode_value = df2['listing_name'].mode()[0]
df2['listing_name'].fillna(mode_value, inplace=True)
mode_value = df2['host_name'].mode()[0]
df2['host_name'].fillna(mode_value, inplace=True)
```

```python
df2.isnull().sum()
```

```python
df2['last_review_date']
```

```python
median_value=df2['reviews_per_month'].median()
df2['reviews_per_month'].fillna(median_value,inplace=True)
```

```python
df2.isnull().sum()
```

```python
df2[['review_day','review_month', 'review_year']]=df2['last_review_date'].str.split('-', expand =True)
df2.drop('last_review_date', axis=1,inplace=True)
df2.head()
```

```python
df2['review_year'].value_counts()
```

```python
df2 = df2.dropna()

df2['review_day']=df2['review_day'].astype(int)
df2['review_year']=df2['review_year'].astype(int)
df2['review_month']=df2['review_month'].astype(int)
```

```python
mean_value=int(np.mean(df2['review_day']))
df2['review_day'].fillna(mean_value,inplace=True)
mean_value=int(np.mean(df2['review_year']))
df2['review_year'].fillna(mean_value,inplace=True)

mean_value=int(np.mean(df2['review_month']))
df2['review_year'].fillna(mean_value,inplace=True)
df2.isnull().sum()
```

```python
df2.plot(subplots=True)
```

```python
frame=df2[['price (in dollars)','minimum_stay','no_of_reviews']]

frame.set_index('minimum_stay',inplace=True)
frame.reset_index(inplace=True)

frame.plot(kind='scatter', x='no_of_reviews', y='price (in dollars)')
plt.title('Scatter Plot of Price vs no_of_reviews')
plt.xlabel('no_of_reviews')
plt.ylabel('price (in dollars)')
plt.show()
```
